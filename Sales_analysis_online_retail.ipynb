**Customer Segmentation by Sales:** Using the calculated total sales to segment customers based on their purchasing behavior. To identify high-value customers, regular customers, and occasional buyers.

```python
# calculating sales per customer 

# Group by CustomerID and sum total sales
customer_total_sales = df_cust_clean.groupby('CustomerID')['TotalSales'].sum()

# Reset index to convert the result into a DataFrame
customer_total_sales = customer_total_sales.reset_index()

# Rename columns for clarity
customer_total_sales.columns = ['CustomerID', 'TotalSales']

display(customer_total_sales.head(10))
```

```python
from sklearn.cluster import KMeans

# Number of clusters
num_clusters = 3

# Fit k-means clustering on total sales
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
customer_total_sales['Cluster'] = kmeans.fit_predict(customer_total_sales[['TotalSales']])
```

```python
# Create a scatter plot of customer segments
plt.figure(figsize=(10, 6))

# Plot high-value customers with 'o' marker
sns.scatterplot(x='CustomerID', y='TotalSales', data=customer_total_sales[customer_total_sales['Cluster'] == 0], marker='o', color='red', label='High-Value')

# Plot regular customers with 's' marker
sns.scatterplot(x='CustomerID', y='TotalSales', data=customer_total_sales[customer_total_sales['Cluster'] == 1], marker='s', color='blue', label='Regular')

# Plot one-time buying customers with '^' marker
sns.scatterplot(x='CustomerID', y='TotalSales', data=customer_total_sales[customer_total_sales['Cluster'] == 2], marker='^', color='green', label='One-Time')

# Set labels and title
plt.xlabel('Customer ID')
plt.ylabel('Total Sales')
plt.title('Customer Segmentation by Sales')

# Show plot
plt.tight_layout()
plt.legend(title='Segment')
plt.show()
```

**calculate and visualizes the sales growth rate for each country between the years 2010 and 2011**

```python
# Group by 'Country', 'invoiceYear', and calculate total sales
country_yearly_sales = df_cust_clean.groupby(['Country', 'invoiceYear'])['TotalSales'].sum().reset_index()

# Pivot the data to have 'Country' as rows and 'invoiceYear' as columns
pivot_country_sales = country_yearly_sales.pivot(index='Country', columns='invoiceYear', values='TotalSales')

# Calculate the sales growth rate for each country
sales_growth_rate = {}
for country in pivot_country_sales.index:
    try:
        start_sales = pivot_country_sales.loc[country, 2010]
        end_sales = pivot_country_sales.loc[country, 2011]
        growth_rate = ((end_sales - start_sales) / start_sales) * 100
        sales_growth_rate[country] = growth_rate
    except KeyError:
        continue

# print(sales_matrix.columns) 
# print(sales_growth_rate)
    

```

```python
# Create a bar plot to visualize sales growth rates by country
plt.figure(figsize=(12, 8))
sns.barplot(x=list(sales_growth_rate.keys()), y=list(sales_growth_rate.values()), color='blue')

# Set plot labels and title
plt.xlabel('Country')
plt.ylabel('Sales Growth Rate (%)')
plt.title('Sales Growth Rate by Country')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90, ha='right')

# Show the plot
plt.tight_layout()
```

**Average Order Value Analysis:**

```python
# Calculate total sales per invoice
invoice_total_sales = df_filter.groupby('InvoiceNo')['TotalSales'].sum()

# Calculate total quantity per invoice
invoice_total_quantity = df_filter.groupby('InvoiceNo')['Quantity'].sum()

# Calculate average order value (total sales divided by total quantity)
average_order_value = invoice_total_sales / invoice_total_quantity

# Add average order value back to the dataframe
df_filter['AverageOrderValue'] = df_filter['InvoiceNo'].map(average_order_value)

# Group by 'invoiceYear' and 'invoiceWeekNumber', calculate average order value
average_order_value_by_time = df_filter.groupby(['invoiceYear', 'invoiceWeekNumber'])['AverageOrderValue'].mean()

# Reset index for visualization
average_order_value_by_time = average_order_value_by_time.reset_index()

# Create line plot to visualize average order value over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='invoiceWeekNumber', y='AverageOrderValue', hue='invoiceYear', data=average_order_value_by_time)
plt.xlabel('Invoice Week Number')
plt.ylabel('Average Order Value')
plt.title('Average Order Value Analysis Over Time')
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

 **Average Order Value Analysis:**
   Calculating the average order value (total sales divided by the number of orders) and analyze how it changes over time or across different customer segments.


```python
# Calculate total sales per invoice
invoice_total_sales = cluster_df.groupby('InvoiceNo')['TotalSales_x'].sum()

# Calculate total quantity per invoice
invoice_total_quantity = cluster_df.groupby('InvoiceNo')['Quantity'].sum()

# Calculate average order value (total sales divided by total quantity)
average_order_value = invoice_total_sales / invoice_total_quantity

# Merge the average_order_value back to the cluster_df DataFrame
cluster_df['AverageOrderValue'] = cluster_df['InvoiceNo'].map(average_order_value)

# Group by 'Cluster' and calculate average order value for each segment
average_order_value_by_segment = cluster_df.groupby('Cluster')['AverageOrderValue'].mean()

# Reset index for visualization
average_order_value_by_segment = average_order_value_by_segment.reset_index()

# Map numerical segment IDs to labels
segment_labels = {
    0: 'High-Value',
    1: 'Regular',
    2: 'One-Time'
}
average_order_value_by_segment['SegmentLabel'] = average_order_value_by_segment['Cluster'].map(segment_labels)

# Create a bar plot to visualize average order value by segment
plt.figure(figsize=(10, 6))
sns.barplot(x='SegmentLabel', y='AverageOrderValue', data=average_order_value_by_segment, palette='viridis')
plt.xlabel('Customer Segment')
plt.ylabel('Average Order Value')
plt.title('Average Order Value Analysis by Customer Segment')
plt.tight_layout()
plt.show()
```
